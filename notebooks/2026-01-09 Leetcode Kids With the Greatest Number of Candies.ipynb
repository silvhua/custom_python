There are `n` kids with candies. You are given an integer array `candies`, where each `candies[i]` represents the number of candies the $ i^{th} $ kid has, and an integer `extraCandies`, denoting the number of extra candies that you have.

Return a boolean array `result` of length `n`, where `result[i]` is `true` if, after giving the $ i^{th} $ kid all the `extraCandies`, they will have the greatest number of candies among all the kids, or `false` otherwise.

Note that multiple kids can have the greatest number of candies.

 

Example 1:
```
Input: candies = [2,3,5,1,3], extraCandies = 3
Output: [true,true,true,false,true] 
Explanation: If you give all extraCandies to:
- Kid 1, they will have 2 + 3 = 5 candies, which is the greatest among the kids.
- Kid 2, they will have 3 + 3 = 6 candies, which is the greatest among the kids.
- Kid 3, they will have 5 + 3 = 8 candies, which is the greatest among the kids.
- Kid 4, they will have 1 + 3 = 4 candies, which is not the greatest among the kids.
- Kid 5, they will have 3 + 3 = 6 candies, which is the greatest among the kids.
```
Example 2:
```
Input: candies = [4,2,1,1,2], extraCandies = 1
Output: [true,false,false,false,false] 
Explanation: There is only 1 extra candy.
Kid 1 will always have the greatest number of candies, even if a different kid is given the extra candy.
```
Example 3:
```
Input: candies = [12,1,12], extraCandies = 10
Output: [true,false,true]
```

Constraints:

* `n == candies.length`
* `2 <= n <= 100`
* `1 <= candies[i] <= 100`
* `1 <= extraCandies <= 50`

## Skeleton code
```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
```

## Iteration 1
* Time: $ O(n^2) $
* Space: O(2n)

````python
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
        print(f"candies: {candies}\textraCandies: {extraCandies}")
        result = candies.copy()
        for i in range(len(candies)):
            print(f"\ti: {i}")
            updated = candies.copy()
            updated_candy_for_i = candies[i] + extraCandies
            updated[i] = updated_candy_for_i
            for j in range(len(candies)):
                is_greater_amount = updated_candy_for_i >= candies[j] 
                print(f"\t\tj: {j}; comparison result: {is_greater_amount}")
                if is_greater_amount == False:
                    break
            result[i] = is_greater_amount
        print(f"\tResult: {result}")
        return result


cases = {
    1: {
        "candies": [2,3,5,1,3],
        "extraCandies": 3,
        "Output": [True,True,True,False,True]
    },
    2: {
        "candies": [4,2,1,1,2], 
        "extraCandies": 1,
        "Output": [True,False,False,False,False] 
    },
    3: {
        "candies": [12,1,12], "extraCandies": 10,
        "Output": [True,False,True]
    }
}

solution = Solution()

for case in cases:
    print(f"*** Case {case} ***")
    candies = cases[case]["candies"]
    extraCandies = cases[case]["extraCandies"]
    Output = cases[case]["Output"]
    result = solution.kidsWithCandies(candies, extraCandies)
    print(f"Answer is correct: {result == Output}\n\n")
````
