# Maximum Average Subarray I
* No NeetCode link
* [LeetCode link](https://leetcode.com/problems/maximum-average-subarray-i/description/)

You are given an integer array `nums` consisting of `n` elements, and an integer `k`.

Find a contiguous subarray whose length is equal to `k` that has the maximum average value and return this value. Any answer with a calculation error less than $ 10^{-5} $ will be accepted.

 

Example 1:
```
Input: nums = [1,12,-5,-6,50,3], k = 4
Output: 12.75000
Explanation: Maximum average is (12 - 5 - 6 + 50) / 4 = 51 / 4 = 12.75
```
Example 2:
```
Input: nums = [5], k = 1
Output: 5.00000
 ```

Constraints:

* `n` == `nums.length`
* 1 <= `k` <= `n` <= $ 10^5 $
* $ -10^4 $ <= `nums[i]` <= $ 10^4 $

## Skeleton code
```python
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
```

## Iteration 1
* Time: $ O(n) $
* Space: $ O(k) $

````python
class Solution:
    def findMaxAverage(self, nums, k) -> float:
        print(f"nums: {nums};\tk: {k}")
        max_mean = float('-inf')
        L, R = 0, k
        while R <= len(nums):
            subarray = nums[L:R]
            current_mean = sum(subarray) / int(k)
            print(f"\tCurrent subarray: {subarray}\n\t\tMean: {current_mean}")
            max_mean = max(max_mean, current_mean)
            L += 1
            R += 1
        print(f"\t** Max mean: {max_mean}")
        return max_mean


cases = {
    1: {
        "nums": [1,12,-5,-6,50,3], 
        "k": 4,
        "Output": 12.75000
    },
    2: {
        "nums": [5], 
        "k": 1,
        "Output": 5.00000
    }
}

solution = Solution()

for case in cases:
    print(f"*** Case {case} ***\n")
    
    nums = cases[case]["nums"]
    k = cases[case]["k"]
    Output = cases[case]["Output"]
    result = solution.findMaxAverage(nums, k)
    print(f"Answer is correct: {result == Output}\n\n")
````

## Iteration 5 [Accepted]
Result: Accepted
* Time: $ O(n) $

````python
class Solution:
    def findMaxAverage(self, nums, k) -> float:
        L, R = 0, k
        subarray = nums[L: R]
        current_sum = sum(subarray)
        current_mean = current_sum / k
        max_mean = current_mean
        for i in range(R, len(nums)):
            current_sum -= nums[L]
            current_sum += nums[i]
            current_mean = current_sum / k
            # print(f"L: {L}\tR: {i}\tsum: {current_sum}")
            max_mean = max(max_mean, current_mean)
            L += 1
        return max_mean

cases = {
    1: {
        "nums": [1,12,-5,-6,50,3], 
        "k": 4,
        "Output": 12.75000
    },
    2: {
        "nums": [5], 
        "k": 1,
        "Output": 5.00000
    },
    3: {
        "nums": [0,1,1,3,3], 
        "k": 4,
        "Output": 2.00000
    }
}

solution = Solution()

for case in cases:
    print(f"*** Case {case} ***\n")
    
    nums = cases[case]["nums"]
    k = cases[case]["k"]
    Output = cases[case]["Output"]
    result = solution.findMaxAverage(nums, k)
    print(f"Result: {result}\tExpected: {Output}")
    print(f"Answer is correct: {result == Output}\n\n")
````
