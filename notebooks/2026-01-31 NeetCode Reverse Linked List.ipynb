* [NeetCode link](https://neetcode.io/problems/reverse-a-linked-list/question?list=allNC)
* [LeetCode link](https://leetcode.com/problems/reverse-linked-list/)

Given the beginning of a singly linked list `head`, reverse the list, and return the new beginning of the list.

Example 1:
```
Input: head = [0,1,2,3]

Output: [3,2,1,0]
```
Example 2:
```
Input: head = []

Output: []
```
Constraints:

* `0 <= The length of the list <= 1000`.
* `-1000 <= Node.val <= 1000`

## Solutions for handling duplicate values in linked list creation

````python
# Solution 1: Use list instead of dictionary to store nodes
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head):
        if head:
            self.reverse_pointer(head)
        else:
            self.new_head = None
        return self.new_head
    
    def createLinkedList(self, node_values):
        if not node_values:
            return None
        
        # Create all nodes first (allows duplicates)
        nodes = [ListNode(val) for val in node_values]
        
        # Link them together
        for i in range(len(nodes) - 1):
            nodes[i].next = nodes[i + 1]
        
        # Store nodes for testing purposes (using index as key instead of value)
        self.nodes = {i: node for i, node in enumerate(nodes)}
        
        return nodes[0]  # Return head
    
    def reverse_pointer(self, node, previous_node = None):
        updated_node = ListNode(node.val, previous_node)
        if node.next:
            new_node = self.reverse_pointer(node.next, updated_node)
        else:
            if previous_node and previous_node.next:
                node.next = ListNode(previous_node.next, previous_node.next.next)
            self.new_head = updated_node
        return updated_node

# Test with duplicate values
cases = {
    1: {
        "head": [0,1,2,3],
        "Output": [3,2,1,0]
    },
    2: {
        "head": [],
        "Output": []
    },
    3: {
        "head": [1,1,2,1],  # Test with duplicates
        "Output": [1,2,1,1]
    }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    node_values = case["head"]
    head = solution.createLinkedList(node_values)
        
    Output = case["Output"]
    new_head = solution.reverseList(head)
    
    # Convert linked list back to list for comparison
    result = []
    current = new_head
    while current:
        result.append(current.val)
        current = current.next
    
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````

````python
# Solution 2: Simple iterative approach (most efficient)
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head):
        if head:
            self.reverse_pointer(head)
        else:
            self.new_head = None
        return self.new_head
    
    def createLinkedList(self, node_values):
        if not node_values:
            return None
        
        head = ListNode(node_values[0])
        current = head
        
        # Create and link nodes one by one
        for val in node_values[1:]:
            current.next = ListNode(val)
            current = current.next
        
        return head
    
    def reverse_pointer(self, node, previous_node = None):
        updated_node = ListNode(node.val, previous_node)
        if node.next:
            new_node = self.reverse_pointer(node.next, updated_node)
        else:
            if previous_node and previous_node.next:
                node.next = ListNode(previous_node.next, previous_node.next.next)
            self.new_head = updated_node
        return updated_node

# Helper function to convert linked list to list for testing
def linked_list_to_list(head):
    result = []
    current = head
    while current:
        result.append(current.val)
        current = current.next
    return result

# Test with duplicate values
cases = {
    1: {
        "head": [0,1,2,3],
        "Output": [3,2,1,0]
    },
    2: {
        "head": [],
        "Output": []
    },
    3: {
        "head": [1,1,2,1],  # Test with duplicates
        "Output": [1,2,1,1]
    },
    4: {
        "head": [5,5,5],  # All same values
        "Output": [5,5,5]
    }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    node_values = case["head"]
    head = solution.createLinkedList(node_values)
    
    print(f"Original linked list: {linked_list_to_list(head)}")
        
    Output = case["Output"]
    new_head = solution.reverseList(head)
    result = linked_list_to_list(new_head)
    
    correct_for_case = result == Output
    print(f"Reversed result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
