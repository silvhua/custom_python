You are given an integer array `nums`, move all `0`'s to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

Example 1:
```
Input: nums = [0,0,1,2,0,5]

Output: [1,2,5,0,0,0]
```
Example 2:
```
Input: nums = [0,1,0]

Output: [1,0,0]
```
Constraints:

* 1 <= nums.length <= 10,000
* $ -2^{31} $ <= nums[i] <= $ 2^{31} $ - 1

Follow up: Could you minimize the total number of operations done?


## Skeleton code
```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
```

## Iteration 1
* Time: $ O() $
* Space: $ O() $

````python


class Solution:
    def moveZeroes(self, nums) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        index = 0
        max_index = len(nums)
        while index < max_index:
            num = nums[index]
            if num == 0:
                nums = nums[:index] + nums[index + 1:]
                nums.append(0)
                max_index -= 1 
            else:
                index += 1
            print(f"\tindex: {index}\tnum: {num}\tnums: {nums}")
        return nums
                



cases = {
    1: {
        "nums": [0,0,1,2,0,5],
        "Output": [1,2,5,0,0,0]
    },
    2: {
        "nums": [0,1,0],
        "Output": [1,0,0]
    },
    3: {
        "nums": [0,1,0,3,12],
        "Output": [1,3,12,0,0]
    }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    nums = case["nums"]
    Output = case["Output"]
    result = solution.moveZeroes(nums)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
    print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
