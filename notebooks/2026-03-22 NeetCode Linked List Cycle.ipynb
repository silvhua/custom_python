# Linked List Cycle
* [LeetCode link](https://leetcode.com/problems/linked-list-cycle/)
* [NeetCode link](https://neetcode.io/problems/linked-list-cycle-detection/question?list=neetcode150)

Given `head`, the head of a linked list, determine if the linked list has a cycle in it.

There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the `next` pointer. Internally, `pos` is used to denote the index of the node that tail's `next` pointer is connected to. Note that `pos` is not passed as a parameter.

Return `true` if there is a cycle in the linked list. Otherwise, return `false`.

 
Example 1:
```
Input: head = [3,2,0,-4], pos = 1
Output: true
```
Explanation: There is a cycle in the linked list, where the tail connects to the 1st node (0-indexed).

Example 2:
```
Input: head = [1,2], pos = 0
Output: true
```
Explanation: There is a cycle in the linked list, where the tail connects to the 0th node.

Example 3:
```
Input: head = [1], pos = -1
Output: false
```
Explanation: There is no cycle in the linked list.
 

Constraints:

* The number of the nodes in the list is in the range [0, $ 10^4 $].
* $ -10^5 $ <= `Node.val` <= $ 10^5 $
* `pos` is `-1` or a valid index in the linked-list.
 

Follow up: Can you solve it using $ O(1) $ (i.e. constant) memory?

## Iteration 1
* Time: $ O() $
* Space: $ O() $

````python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def hasCycle(self, head) -> bool:
        slow, fast = head, head
        i = 0
        while fast and fast.next:
            if slow.val == fast.val:
                print(f"i: {i}\tslow.val: {slow.val}\tfast.val: {fast.val}")
                return True
            slow = slow.next
            fast = fast.next.next
            i += 1
        return False

    def createLinkedList(self, node_values, pos=-1):
        self.nodes = {}
        if len(node_values) > 0:
            self.nodes[len(node_values) - 1] = ListNode(node_values[-1])
            for i in range(len(node_values) - 2, -1, -1):
                val = node_values[i]
                next_val = node_values[i + 1]
                self.nodes[i] = ListNode(val, self.nodes[i + 1])
            head = self.nodes[0]
            if pos >= 0: # If there is a cycle, point the tail to the head of the cyle
                self.nodes[len(node_values) - 1].next = self.nodes[pos]
        else:
            head = ListNode(None)
        return head

cases = {
    # 1: {
    #     "head": [3,2,0,-4], 
    #     "pos": 1,
    #     "Output": True
    # },
    # 2: {
    #     "head": [1,2], 
    #     "pos": 0,
    #     "Output": True
    # },
    # 3: {
    #     "head": [1], 
    #     "pos": -1,
    #     "Output": False
    # },
    # 4: {
    #     "head": [1,2,3,4], 
    #     "pos": 1,
    #     "Output": True
    # },
    5: {
        "head": [1,2], 
        "pos": -1,
        "Output": False
    }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    node_values = case["head"]
    head = solution.createLinkedList(node_values)
    pos = case["pos"]
    Output = case["Output"]
    result = solution.hasCycle(head)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
