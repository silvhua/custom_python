# Combination Sum III
* [LeetCode link](https://leetcode.com/problems/combination-sum-iii/)

Find all valid combinations of `k` numbers that sum up to `n` such that the following conditions are true:

* Only numbers 1 through 9 are used.
* Each number is used at most once.
Return a list of all possible valid combinations. The list must not contain the same combination twice, and the combinations may be returned in any order.


Example 1:
```
Input: k = 3, n = 7
Output: [[1,2,4]]
```
Explanation:
```
1 + 2 + 4 = 7
```
There are no other valid combinations.
Example 2:
```
Input: k = 3, n = 9
Output: [[1,2,6],[1,3,5],[2,3,4]]
```
Explanation:
```
1 + 2 + 6 = 9
1 + 3 + 5 = 9
2 + 3 + 4 = 9
```
There are no other valid combinations.
Example 3:
```
Input: k = 4, n = 1
Output: []
```
Explanation: There are no valid combinations.
Using 4 different numbers in the range [1,9], the smallest sum we can get is 1+2+3+4 = 10 and since 10 > 1, there are no valid combination.
 
Constraints:
* 2 <= `k` <= 9
* 1 <= `n` <= 60

## Iteration 1
* Time: $ O() $
* Space: $ O() $

````python
class Solution(object):
    def combinationSum3(self, k, n):
        """
        :type k: int
        :type n: int
        :rtype: List[List[int]]
        """
        self.n = n
        self.k = k

        self.result = []
        for max_digit in range(min(9, n), 0, -1):
            print(f"\tinterim result: {self.result}")
            available_digits = [x for x in range(1, max_digit + 1)]
            if self.filterDigit(max_digit, available_digits):
                print(f"max digit: {max_digit}")
                self.getRemainder(max_digit, available_digits, [])
        return self.result
    
    def filterDigit(self, digit, available_digits):
        """
        Filter out digits that are too large.
        """
        lowest_digits = available_digits[:self.k - 1]
        sum_of_lowest_digits = sum(lowest_digits)
        # print(f"\tlowest digits: {lowest_digits}\tsum: {sum_of_lowest_digits}")
        return sum_of_lowest_digits + digit <= self.n


    def getRemainder(self, digit, available_digits, digits, sum=0, n_digits=0):
        print(f"\tdigit: {digit}\tavailable digits: {available_digits}")
        while n_digits <= self.k and sum <= self.n and len(available_digits) > 0 and digit in available_digits:
            available_digits.remove(digit)
            remainder = self.n - sum - digit
            if remainder > 0:
                sum += digit
                n_digits += 1
                print(f"\t\tsum: {sum}")
                digits.append(digit)
                print(f"\t\t\tavailable digits: {available_digits}\tsum: {sum}")
                self.getRemainder(remainder, available_digits, digits, sum, n_digits)
            elif remainder == 0 and self.k == n_digits + 1:
                print(f"\t\tsuccess!")
                digits.append(digit)
                sum += digit
                n_digits += 1
            elif self.filterDigit(digit - 1, available_digits):
                self.getRemainder(digit - 1, available_digits, digits, sum, n_digits)
        if sum == self.n and n_digits == self.k:
            print(f"\tFound option: {digits}")
            self.result.append(sorted(digits))


cases = {
    1: {
        "k": 3, 
        "n": 7,
        "Output": [[1,2,4]]
    },
    # 2: {
    #     "k": 3, 
    #     "n": 9,
    #     "Output": [[1,2,6],[1,3,5],[2,3,4]]
    # },
    # 3: {
    #     "k": 4, 
    #     "n": 1,
    #     "Output": []
    # }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    k = case["k"]
    n = case["n"]
    Output = case["Output"]
    result = solution.combinationSum3(k, n)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
