# Maximum Level Sum of a Binary Tree
* [LeetCode link](https://leetcode.com/problems/maximum-level-sum-of-a-binary-tree/)

Given the `root` of a binary tree, the level of its root is 1, the level of its children is 2, and so on.

Return the smallest level `x` such that the sum of all the values of nodes at level `x` is maximal.

Example 1:
```
Input: root = [1,7,0,7,-8,null,null]
Output: 2
```
Explanation: 
Level 1 sum = 1.
Level 2 sum = 7 + 0 = 7.
Level 3 sum = 7 + -8 = -1.
So we return the level with the maximum sum which is level 2.
Example 2:
```
Input: root = [989,null,10250,98693,-89388,null,null,null,-32127]
Output: 2
```

Constraints:

* The number of nodes in the tree is in the range [1, $ 10^4 $].
* $-10^5 $ <= Node.val <= $ 10^5 $

## Iteration 1
* Time: $ O(N) $
* Space: $ O(N) $

````python
# Definition for a binary tree node.
from typing import Deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def maxLevelSum(self, root) -> int:
        self.bfs(root)
        max_level_sum = max(self.level_sums, key=self.level_sums.get)
        return max_level_sum

    def bfs(self, root):
        queue = Deque()
        self.level_sums = {}

        if root:
            queue.append(root)
            self.level_sums[1] = root.val
    
        level = 2
        while len(queue) > 0:
            level_sum = 0
            for i in range(len(queue)):
                curr = queue.popleft()
                if curr.left:
                    queue.append(curr.left)
                    if curr.left.val:
                        level_sum += curr.left.val
                if curr.right:
                    queue.append(curr.right)
                    if curr.right.val:
                        level_sum += curr.right.val
                self.level_sums[level] = level_sum
            level += 1

    def printLevelSums(self):
        print(f"\tlevel sums:")
        for key, value in self.level_sums.items():
            print(f'\t\t{key}: {value}')

    def buildTree(self, nodes):
        # Treat the tree as a heap
        self.nodes = [None] + nodes
        self.tree = self.createNode(1)
        return self.tree


    def createNode(self, index):
        if index >= len(self.nodes):
            return None
        left_child_index = index * 2
        right_child_index = index * 2 + 1
        val = self.nodes[index]
        # print(f"\tval: {val}\tindex: {index}\tleft child index: {left_child_index}\tright child index: {right_child_index}")
        node = TreeNode(
            val,
            self.createNode(left_child_index),
            self.createNode(right_child_index)
            )
        return node
    
    def printNodes(self, root, nodes_list = []):
        if not root:
            nodes_list.append(None)
            return
        nodes_list.append(root.val)
        self.addToNodesList(root, nodes_list)
        print(f"Tree nodes: {[val for val in nodes_list]}")

    def addToNodesList(self, node, nodes_list):
        if not node:
            return
        if node.left:
            nodes_list.append(node.left.val)
        if node.right:
            nodes_list.append(node.right.val)
        self.addToNodesList(node.left, nodes_list)
        self.addToNodesList(node.right, nodes_list)


cases = {
    1: {
        "root": [1,7,0,7,-8,None,None],
        "Output": 2
    },
    2: {
        "root": [989,None,10250,98693,-89388,None,None,None,-32127],
        "Output": 2
    },
    # 3: {
    #     "Output": 
    # }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    root = case["root"]
    Output = case["Output"]

    tree = solution.buildTree(root)
    solution.printNodes(tree, [])

    result = solution.maxLevelSum(tree)
    solution.printLevelSums()
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
