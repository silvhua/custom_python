# Maximum Depth of Binary Tree
* [LeetCode link](https://leetcode.com/problems/maximum-depth-of-binary-tree/)
* [NeetCode link](https://neetcode.io/problems/depth-of-binary-tree/question?list=allNC)

Given the `root` of a binary tree, return its depth.

The depth of a binary tree is defined as the number of nodes along the longest path from the root node down to the farthest leaf node.

Example 1:
```
Input: root = [1,2,3,null,null,4]

Output: 3
```
Example 2:
```
Input: root = []

Output: 0
```
Constraints:

* 0 <= The number of nodes in the tree <= 100.
* -100 <= `Node.val` <= 100

## Iteration 1 solve the problem
* Time: $ O(n) $
* Space: $ O(log(n)) $

* Runtime 3 ms Beats 27.02%
* Memory 20.39 MB Beats 38.65%

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def maxDepth(self, root) -> int:
        max_length = self.preorder_length(root, 0)
        return max_length

    def preorder_length(self, root, length):
        if not root:
            return length
        length += 1
        left_length = self.preorder_length(root.left, length)
        right_length = self.preorder_length(root.right, length)
        return max(right_length, left_length)

    def buildTree(self, nodes):
        # Treat the tree as a heap
        self.nodes = [None] + nodes
        self.tree = self.createNode(1)
        return self.tree


    def createNode(self, index):
        if index >= len(self.nodes):
            return None
        left_child_index = index * 2
        right_child_index = index * 2 + 1
        val = self.nodes[index]
        # print(f"\tval: {val}\tindex: {index}\tleft child index: {left_child_index}\tright child index: {right_child_index}")
        node = TreeNode(
            val,
            self.createNode(left_child_index),
            self.createNode(right_child_index)
            )
        return node
    
    def printNodes(self, root, nodes_list):
        if not root:
            nodes_list.append(None)
            return
        nodes_list.append(root.val)
        self.addToNodesList(root, nodes_list)
        print(f"Tree nodes: {[val for val in nodes_list]}")

    def addToNodesList(self, node, nodes_list):
        
        if not node:
            return
        if node.left:
            nodes_list.append(node.left.val)
        if node.right:
            nodes_list.append(node.right.val)
        self.addToNodesList(node.left, nodes_list)
        self.addToNodesList(node.right, nodes_list)
        
cases = {
    1: {
        "root": [1,2,3,None,None,4],
        "Output": 3
    },
    2: {
        "root": [],
        "Output": 0
    },
    3: {
        "root": [3,9,20,None,None,15,7],
        "Output": 3
    },
    4: {
        "root": [1,None,2],
        "Output": 2
    }
}

correct = []

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    
    nodes = case["root"]
    solution = Solution()
    tree = solution.buildTree(nodes)
    solution.printNodes(tree, [])

    Output = case["Output"]

    result = solution.maxDepth(tree)

    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}")
    print(f"\n")
print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
