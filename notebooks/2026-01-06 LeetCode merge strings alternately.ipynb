# Merge Strings Alternately
* [NeetCode link](https://neetcode.io/problems/merge-strings-alternately/question?list=allNC)
* [LeetCode link](https://leetcode.com/problems/merge-strings-alternately/)

You are given two strings, word1 and word2. Construct a new string by merging them in alternating order, starting with word1 â€” take one character from word1, then one from word2, and repeat this process.

If one string is longer than the other, append the remaining characters from the longer string to the end of the merged result.

Return the final merged string.

Example 1:
```
Input: word1 = "abc", word2 = "xyz"

Output: "axbycz"
```
Example 2:
```
Input: word1 = "ab", word2 = "abbxxc"

Output: "aabbbxxc"
```
Constraints:

1 <= word1.length, word2.length <= 100
word1 and word2 consist of lowercase English letters.


## Skeleton code
```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
```

## Iteration 1 [CORRECT]
* Time: O(n)
* Space: O(n)

````python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        max_length = max(len(word1), len(word2))
        print(f"Max word length: {max_length}")
        word = ""
        for i in range(max_length):
            if i < len(word1):
                word += word1[i]
            if i < len(word2):
                word += word2[i]
        print(f"Merged word: {word}")
        return word


word1 = "ab"
word2 = "abbxxc"
solution = Solution()
solution.mergeAlternately(word1, word2)
````
