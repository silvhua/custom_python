# Find the Difference of Two Arrays
* [NeetCode link](https://neetcode.io/problems/find-the-difference-of-two-arrays/question?list=allNC)
* [LeetCode link](https://leetcode.com/problems/find-the-difference-of-two-arrays/)

You are given two 0-indexed integer arrays `nums1` and `nums2`, return a list `answer` of size 2 where:

`answer[0]` is a list of all distinct integers in `nums1` which are not present in `nums2`.
`answer[1]` is a list of all distinct integers in `nums2` which are not present in `nums1`.
Note that the integers in the lists may be returned in any order.

Example 1:
```
Input: nums1 = [1,2,3], nums2 = [2,4,6]

Output: [[1,3],[4,6]]
```
Example 2:
```
Input: nums1 = [1,2,3,3], nums2 = [1,1,2,2]

Output: [[3],[]]
```
Constraints:

* `1 <= nums1.length, nums2.length <= 1000`
* `-1000 <= nums1[i], nums2[i] <= 1000`

## Skeleton code
```python
class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
```

## Iteration 1
* Time: $ O(n+m) $
* Space: $ O(n+m) $

````python
class Solution:
    def findDifference(self, nums1, nums2):
        longest_array = max(len(nums1), len(nums2))
        answer0 = []
        answer1 = []
        len1 = len(nums1)
        len2 = len(nums2)
        for i in range(longest_array):
            if i < len1:
                num1 = nums1[i]
                if num1 not in nums2 and num1 not in answer0:
                    answer0.append(num1)
            if i < len2:
                num2 = nums2[i]
                if num2 not in nums1 and num2 not in answer1:
                    answer1.append(num2)
        return [answer0, answer1]



cases = {
    1: {
        "nums1": [1,2,3], 
        "nums2": [2,4,6],
        "Output": [[1,3],[4,6]]
    },
    2: {
        "nums1": [1,2,3,3], 
        "nums2": [1,1,2,2],
        "Output": [[3],[]]
    },
    # 3: {
    #     "Output": 
    # }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    nums1 = case["nums1"]
    nums2 = case["nums2"]
    Output = case["Output"]
    result = solution.findDifference(nums1, nums2)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
    print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````

* Time: $ O(n+m) $
* Space: $ O(n+m) $

````python
class Solution:
    def findDifference(self, nums1, nums2):
        nums1 = set(nums1)
        nums2 = set(nums2)
        answer0 = list(nums1 - nums2)
        answer1 = list(nums2 - nums1)
        return [answer0, answer1]



cases = {
    1: {
        "nums1": [1,2,3], 
        "nums2": [2,4,6],
        "Output": [[1,3],[4,6]]
    },
    2: {
        "nums1": [1,2,3,3], 
        "nums2": [1,1,2,2],
        "Output": [[3],[]]
    },
    # 3: {
    #     "Output": 
    # }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    nums1 = case["nums1"]
    nums2 = case["nums2"]
    Output = case["Output"]
    result = solution.findDifference(nums1, nums2)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
    print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
