# Find Pivot Index
* [NeetCode link](https://neetcode.io/solutions/find-pivot-index)
* [LeetCode link](https://leetcode.com/problems/find-pivot-index/)

Given an array of integers `nums`, calculate the pivot index of this array.

The pivot index is the index where the sum of all the numbers strictly to the left of the index is equal to the sum of all the numbers strictly to the index's right.

If the index is on the left edge of the array, then the left sum is `0` because there are no elements to the left. This also applies to the right edge of the array.

Return the leftmost pivot index. If no such index exists, return `-1`.

Example 1:
```
Input: nums = [1,7,3,6,5,6]
Output: 3
Explanation:
The pivot index is 3.
Left sum = nums[0] + nums[1] + nums[2] = 1 + 7 + 3 = 11
Right sum = nums[4] + nums[5] = 5 + 6 = 11
```
Example 2:
```
Input: nums = [1,2,3]
Output: -1
Explanation:
There is no index that satisfies the conditions in the problem statement.
Example 3:
```
```
Input: nums = [2,1,-1]
Output: 0
Explanation:
The pivot index is 0.
Left sum = 0 (no elements to the left of index 0)
Right sum = nums[1] + nums[2] = 1 + -1 = 0
```
Constraints:

* 1 <= `nums.length` <= $ 10^4 $
* -1000 <= `nums[i]` <= 1000

* Time: $ O(2N) $
* Space: $ O(2N) $

````python
# Start by assumiming the pivot is in the most left
class Solution:
    def pivotIndex(self, nums) -> int:
        prefix_sum = [nums[0]]
        postfix_sum = [nums[-1]]
        for i in range(1, len(nums)):
            prefix_sum.append(prefix_sum[-1] + nums[i])
            postfix_sum.append(postfix_sum[-1] + nums[-i - 1])
            print(f"\ti: {i}\tprefix: {prefix_sum}\tpostfix: {postfix_sum}")
        for j in range(len(nums)):
            left = prefix_sum[j]
            right = postfix_sum[len(nums) - j - 1]
            print(f"\tj: {j}\tleft: {left}\tright: {right}")
            if left == right:
                return j
        return -1


cases = {
    1: {
        "nums": [1,7,3,6,5,6],
        "Output": 3
    },
    2: {
        "nums": [1,2,3],
        "Output": -1
    },
    3: {
        "nums": [2,1,-1],
        "Output": 0
    }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    nums = case["nums"]
    Output = case["Output"]
    result = solution.pivotIndex(nums)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
    print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
