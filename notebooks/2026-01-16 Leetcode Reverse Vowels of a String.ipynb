# Reverse Vowels of a String
* [LeetCode link](https://leetcode.com/problems/reverse-vowels-of-a-string/description/)

Given a string `s`, reverse only all the vowels in the string and return it.

The vowels are `'a'`, `'e'`, `'i'`, 'o', and `'u'`, and they can appear in both lower and upper cases, more than once.

 

Example 1:
```
Input: s = "IceCreAm"

Output: "AceCreIm"
```
Explanation:

The vowels in s are ['I', 'e', 'e', 'A']. On reversing the vowels, s becomes "AceCreIm".

Example 2:
```
Input: s = "leetcode"

Output: "leotcede"
```
 

Constraints:

* 1 <= `s.length` <= 3 * $ 10^5 $
* `s` consist of printable ASCII characters.
 

## Skeleton code
```python
class Solution:
    def reverseVowels(self, s: str) -> str:
```

## Iteration 1
* Time: $ O(2N) $
* Space: $ O(3N) $

````python
class Solution:
    def reverseVowels(self, s: str) -> str:
        vowels = ['a', 'e', 'i', 'o', 'u']
        result = []
        vowel_indices = []
        s_vowels = []

        # Collect all the vowels
        for i, char in enumerate(s):
            if s[i].lower() in vowels:
                vowel_indices.append(i)
                s_vowels.append(char)
        vowel_index = -1
        for j, char in enumerate(s):
            if s[j].lower() in vowels:
                result.append(s_vowels[vowel_index])
                vowel_index -= 1
            else:
                result.append(char)
        
        return "".join([char for char in result])




cases = {
    1: {
        "s": "leetcode",
        "Output": "leotcede"
    },
    2: {
        "s": "IceCreAm",
        "Output": "AceCreIm"
    },
    # 3: {
    #     "Output": 
    # }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    s = case["s"]
    Output = case["Output"]
    result = solution.reverseVowels(s)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
    print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````

### Iteration 1.1
* Time: $ O(N/2) $
* Space: $ O(N) $

````python
class Solution:
    def reverseVowels(self, s: str) -> str:
        vowels = ['a', 'e', 'i', 'o', 'u']
        L, R = 0, len(s) - 1
        l_vowel_index, r_vowel_index = None, None
        result = [char for char in s]
        
        while L < R and (l_vowel_index is None or r_vowel_index is None):
            print(f"\tL: {L}\tR: {R}")
            l_char = s[L]
            if l_char.lower() in vowels:
                l_vowel_index = L
            else:
                L += 1
            r_char = s[R]
            if r_char.lower() in vowels:
                r_vowel_index = R
            else:
                R -= 1
            if l_vowel_index != None and r_vowel_index != None:
                result[l_vowel_index] = r_char
                result[r_vowel_index] = l_char
                print(f"\tInterim word: {result}")
                l_vowel_index, r_vowel_index = None, None
                L += 1
                R -= 1

        return "".join([char for char in result])





cases = {
    1: {
        "s": "leetcode",
        "Output": "leotcede"
    },
    2: {
        "s": "IceCreAm",
        "Output": "AceCreIm"
    },
    # 3: {
    #     "Output": 
    # }
}

correct = []
solution = Solution()

for case_number in cases:
    case = cases[case_number]
    print(f"*** Case {case_number} ***")
    for key in case.keys():
        print(f"{key}: {case[key]}", end="\t")
    print("")
    
    s = case["s"]
    Output = case["Output"]
    result = solution.reverseVowels(s)
    correct_for_case = result == Output
    print(f"Result: {result}")
    correct.append(correct_for_case)
    print(f"Answer is correct: {correct_for_case}\n\n")
    print(f"\n***** All cases correct: {sum(correct) == len(correct)} *****")
````
